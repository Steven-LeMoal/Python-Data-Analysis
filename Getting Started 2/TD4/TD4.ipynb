### Exercise 2 : Matplotlib's Object Oriented API

Matplotlib have an object-oriented API used to create more complex figures. Code is a
little more complicated, but the advantage is that we now have full control of where the plot axes. Here‚Äôs a good read when exploring matplotlib

#### 1) Create the same chart as 1.1 using the new method: instantiate figure objects and then call methods or attributes from that object.

```python
x = np.arange(1,5.5,0.5)
y = np.power(x,2)
```

```python
fig = plt.figure()
#plt.style.use(plt.style.available[12])
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title("y en fonction de x")
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()

#fig = plt.figure()
#plt.style.use(plt.style.available[12])
#plt.figure(figsize=(10,6), dpi=80)
#plt.plot(x, y)
#plt.show()
```

#### 2) Create the figure below (the blue curve represents $y=x^2$ while the red curve represents $x=y^2$)

```python
fig = plt.figure()
#plt.style.use(plt.style.available[12])
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.4,0.3,0.3])

ax1.plot(x,y,'b--')
ax1.set_title("y = x^2")
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(y,x,'r--')
ax2.set_title("x = y^2")
ax2.set_xlabel('x')
ax2.set_ylabel('y')

plt.show()
#plt.style.use(plt.style.available[12])
```

#### 4) Create the same subplots you've created in 1.4 using matplotlib API (use `plt.subplots()`

a. Add a title for each plot ‚Äòtitle 1‚Äô, ‚Äòtitle 2‚Äô, etc.

b. Add an x axis and y axis label for each plot (for example, x-axis: X, y-axis: f(x)= x^2

c. Use figsize to make it clearer

d. Show legend on chart

e. Differentiate between each graph using a different color and markers




```python
x = np.arange(1,5.5,0.5)
```

```python
plt.figure(figsize=(20,10))
plt.style.use(plt.style.available[0])

plt.subplot(2,2,1)
plt.title("figure 1")
plt.xlabel('x')
plt.ylabel("f(x)")
plt.plot(x,x**2, label='f(x)')
plt.legend()

plt.subplot(2,2,2)
plt.title("figure 2")
plt.xlabel('x')
plt.ylabel("f(x)")
plt.plot(x,x**3,'b--',label='f(x)')
plt.legend()

plt.subplot(2,2,3)
plt.title("figure 3")
plt.xlabel('x')
plt.ylabel("f(x)")
plt.plot(x,np.log(x),'go',label='f(x)')
plt.legend()

plt.subplot(2,2,4)
plt.title("figure 4")
plt.xlabel('x')
plt.ylabel("f(x)")
plt.plot(x,np.exp(x),'v',label='f(x)')
plt.legend()


plt.show()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(20,10))

axes[0,0].plot(x,x**2, label="square")
axes[0,0].set_title("figure 1")

axes[1,0].plot(x,x**3, label="cube")
axes[1,0].set_title("figure 2")

axes[0,1].plot(x,np.log(x), label="log")
axes[0,1].set_title("figure 3")

axes[1,1].plot(x,np.exp(x), label="exp")
axes[1,1].set_title("figure 4")

for ax in axes.flat:
    ax.legend(loc='upper left')
    
fig.tight_layout()
```
