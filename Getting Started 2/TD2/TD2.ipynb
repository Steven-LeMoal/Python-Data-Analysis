#### Transformation - Use of the monalisa.txt file

10) Create a filter on the image to have a Darker Mona Lisa.

```python
monalisa = np.loadtxt('monalisa.txt')
plt.imshow(monalisa)
```

```python
def brightness(image ,value):
    image = image.astype("int16")
    image = image + value
    return ceil_floor(image)

def ceil_floor(image):
    length = image.shape
    for i in range(length[0]):
        for j in range(length[1]):
            if image[i][j] < 0:
                image[i][j] = 0
            if image[i][j] > 255:
                image[i][j] = 255
    return image
```

```python
plt.imshow(brightness(monalisa,-100))
```

11) Create a vertical gradient from right to left allowing the image to be darker on the left.

```python
grad = [np.linspace(0, 1, x.size, endpoint=True) for x in monalisa]
plt.imshow(monalisa * grad)
```

12) Create a horizontal gradient that will make the Mona Lisa darker on top.

```python
grad = [np.linspace(0, 1, x.size, endpoint=True) for x in np.transpose(monalisa)]
plt.imshow(monalisa * np.transpose(grad))
```
